```python
%matplotlib inline
import audiogen
import matplotlib.pyplot as plt
import numpy as np

a_tone = audiogen.tone(440)
a2_tone = audiogen.tone(446)
a3_tone = audiogen.tone(430)

def add_tones(*tones):
    number_of_tones = len(tones)
    mix_condition = ((audiogen.util.constant(0.5*np.exp(-i)) for i in xrange(number_of_tones)), )
    
    return audiogen.util.mixer(tones, mix_condition)[0]

tones = [audiogen.tone(i) for i in 180*np.arange(1, 4, 1)]
audio_data = add_tones(*tones)

data = list(audiogen.util.crop_with_fades(audio_data, 1, 0.1, 0.1))
plt.plot(data)

audiogen.sampler.play(audiogen.util.crop_with_fades(audio_data, 1, 0.1, 0.1))
```
