```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
from scipy.io import wavfile

sb.set_context("poster")
```

```python
def calculate_spec(data, sampling_frequency):
    # Convert integer to floats between 0 and 1
    tmp_data = data / (2.0 ** 15)
    
    N = len(tmp_data)
    fft = np.abs(np.fft.fft(tmp_data))
    
    number_of_unique_pnts = np.ceil((N+1)/2.0)
    
    spec = fft[0:int(number_of_unique_pnts)]
    
    # Normalize spec
    spec = spec / float(N)
    
    power_spec = 20 * np.log10(spec)
    
    x_axis = np.arange(0, number_of_unique_pnts) * (1.0 * sampling_frequency / N)
    
    return x_axis, power_spec
```

```python
sampling_frequency, sound_data = wavfile.read("audio_generation/data/flute_tone.wav")
sound_left = sound_data[:, 0]
sound_right = sound_data[:, 1]

freq, power_spec = calculate_spec(sound_left, sampling_frequency)
plt.plot(freq, power_spec)
#plt.ylim(-100, -20)
plt.xlim(0, 5000)
```
